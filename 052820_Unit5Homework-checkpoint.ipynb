```python
# extract API keys from .env file

plaid_client_id = os.getenv('PLAID_CLIENT_ID')
plaid_public_key = os.getenv('PLAID_PUBLIC_KEY')
plaid_sbx_secret_key = os.getenv('PLAID_SECRET_SBX_KEY')

# print('client id', type(plaid_client_id))
# print('public key', type(plaid_public_key))
# print('sbx', type(plaid_sbx_secret_key))
```

```python
# create a client object

client = plaid.Client(plaid_client_id, plaid_sbx_secret_key, plaid_public_key,
                      environment = 'sandbox')
```

```python
# set institution id
INSTITUTION_ID = 'ins_109508'
```

```python
# create public token
create_tok_response = client.Sandbox.public_token.create(INSTITUTION_ID, ['transactions', 'income', 'assets'])
print(create_tok_response)
```

```python
# retrieve the access token
exchange_response = client.Item.public_token.exchange(create_tok_response['public_token'])

access_tok = exchange_response['access_token']
print(access_tok)
```

```python
# retrieve accounts
accounts = client.Accounts.get(access_tok)
##pp.pprint(accounts)
```
